```
# 이상기체 상태방정식 계산 예제
# PV = nRT

def ideal_gas_pressure(n, V, T):
    """
    n: 몰수 (mol)
    V: 부피 (m^3)
    T: 온도 (K)
    R: 기체상수 (J/(mol·K))
    """
    R = 8.314  # J/(mol·K)
    P = (n * R * T) / V
    return P

def ideal_gas_volume(n, P, T):
    R = 8.314
    V = (n * R * T) / P
    return V

# 예시 계산
n = 1.0       # mol
V = 0.0224    # m^3 (22.4 L)
T = 273.15    # K (0°C)

P = ideal_gas_pressure(n, V, T)
print(f"압력: {P:.2f} Pa")

# 다른 예시: 압력과 온도로 부피 계산
P2 = 101325   # Pa
T2 = 300      # K
V2 = ideal_gas_volume(n, P2, T2)

print(f"부피: {V2:.4f} m^3")
```
